```python
import numpy as np
import pandas as pd
```

```python
#Loading data
df = pd.read_csv('anime.csv')
df.head()
```

```python
#Checking null values
df.isna().sum()
```

```python
#1.Make a new column for episode count.
def extract_episode(txt):
    check = False
    data = ''
    for i in txt:
        if i == ')':
            check = False
            return data
        if check == True:
            data += i
        if i == '(' :
            check = True
```

```python
df['Episodes']=df['Title'].apply(extract_episode)
df['Episodes']=df['Episodes'].str.replace(" eps" , "")
df['Episodes']=df['Episodes'].astype(int)
```

```python
#Make a new column for timestamp.
def extract_time(txt):
    data = ''
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1, i+20):
                data +=txt[j]
                
            return data
    
```

```python
df['Time_Stamp']=df['Title'].apply(extract_time)
```

```python
#Extract Months from time stamp
def calculate_months(time_str):
   parts = str(time_str).split('-')
   if len(parts) != 2:
       return None
   start_date = pd.to_datetime(parts[0].strip(), format= '%b %Y')
   end_date = pd.to_datetime(parts[1].strip(),format= '%b %Y')
   total_months = (end_date.year- start_date.year) * 12 + (end_date.month - start_date.month)
   return total_months
   df
```

```python
df['Total Months'] = df['Time_Stamp'].apply(calculate_months)
df.head()
```
