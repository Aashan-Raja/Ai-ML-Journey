```python
import numpy as np
import pandas as pd
```

Phase 1 : **Data Cleaning.**

```python
#unclean sample data set
data = {
    'Car_Model' : ['Corrola','Civic','Mehran','Corrola','City','Cultus'],
    'Mileage'   : [50000 , np.nan , 20000 ,50000,80000,15000], #np.nan is an empty data (missing)
    'Price'     : [2500000 , 3000000 , 100000,2500000,500000000,1800000] #50cr is an outlier
}
df = pd.DataFrame(data)
print("--- Orignal Messy Data ---")
data
```

**-Handling Missing Values**

```python
df.isna().sum()
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df
```

**-Handling Duplicates**

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
print('After removing duplicates:\n' , df)
```

**-Outliers Removing**

```python
#Removing those car with price more then 1cr
df = df[df['Price'] < 10000000]
```

```python
print('After Removing outliers:\n',df)
```

Phase 2 : **Data Transformation**
**- Encoding Categorical Data**

**(A)** Label Encoding
**(B)** One-hot-Encoding
**(C)** Target Encoding

```python
#Sample Data Set
data = {
    'car' : ['civic' , 'corrola' , 'mehran' ],
    'condition' : ['good' , 'excellent' , 'bad'],
    'fuel' : ['petrol' , 'diesel' , 'cng']
}
df = pd.DataFrame(data)
df
```

```python
#1 : Label Encoding (using .map function for condition column)
condition_mapping = {'good' : 2 , 'excellent' : 3 , 'bad' :1 }
df['condition'] = df['condition'].map(condition_mapping)
df
```

```python
#2 : One-hot-encoding (used for fuel column using get_dumies)
df = pd.get_dummies(df , columns=['fuel'] , dtype= int)
df
```

**Target Encoding**

```python
# 1.  Data (includes price/Target columns)
data = {
    'Car': ['Civic', 'Corolla', 'Mehran', 'City'],
    'Condition': ['Good', 'Excellent', 'Bad', 'Good'],
    'Fuel': ['Petrol', 'Diesel', 'CNG', 'Petrol'], # Petrol 2 dafa hai
    'Price': [3000000, 4000000, 800000, 2000000]   # Yeh hamara Target hai
}
df = pd.DataFrame(data)

print("--- Original Data ---")
print(df)

# TARGET ENCODING LOGIC


# Step 1:  Calculating each fule type average price
# groupby('Fuel') and ['Price'].mean()
fuel_target_mean = df.groupby('Fuel')['Price'].mean()

# Step 2: mapping average back to fuel column
df['Fuel_Encoded'] = df['Fuel'].map(fuel_target_mean)
print("\n---  After Target Encoding  ---")
print(df[['Car', 'Condition', 'Fuel', 'Fuel_Encoded', 'Price']])
```
