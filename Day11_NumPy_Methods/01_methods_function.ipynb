Generating array using arange 

```python
import numpy as np
arr1d = np.arange(1 , 31)
arr1d
```

Reshaping 1d array into 2d array

```python
arr2d = arr1d.reshape(6,5)
arr2d
```

Checking size , dimension , datatype and no. of cols and rows of array

```python
print(arr2d.shape)
print(arr2d.size)
print(arr2d.ndim)
print(arr2d.dtype)
```

converting one datatype to another

```python
arr3 = arr1d.astype(str)
print(arr3)
```

Aggregation functions (sum,mean,max,min,std,variance)

```python
print(np.mean(arr1d))
print(np.std(arr1d))
print(np.min(arr1d))
print(np.max(arr1d))
print(np.var(arr1d))
print(np.sum(arr1d))
```

Row wise & Column wise sum

```python
print("Row wise")
print(np.sum(arr2d , axis=1))
print("column wise")
print(np.sum(arr2d , axis=0))
```

Printing index of maximum element and minimum element in array

```python
print(arr2d.argmax())
print(arr2d.argmin())
```

Indexing & Slicing

```python
print(arr1d[0:])
print(arr2d)
```

Accessiing (19 , 20 , 24 , 25 ,29 ,30) from arr2d

```python
print(arr2d[3: ,3:])
```

Boolean indexing

```python
arr = np.arange(0 , 11)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
