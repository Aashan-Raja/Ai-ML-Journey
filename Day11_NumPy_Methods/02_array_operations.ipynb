Arthmetic Operations

```python
a1 = np.array([1,2,3,4,5,6,7,8])
a2 = np.array([9,10,11,12,13,14,15,16])
```

```python
add = a1 + a2
add
```

```python
sub = a2 - a1
sub
```

```python
a1 / a2
```

```python
a1 ** a2
```

Broadcasting

```python
l = [1,3,5,6,7]
arr = np.array(l)
arr
```

```python
arr = arr + 10
arr
```

Deep and Shallow copy


```python
#shallow copy
a = np.arange(1,10)
a
```

```python
b = a.copy()
b
```

```python
b = b + 10
b
```

```python
a
```

```python
#deep copy
a = np.arange(0,10)
a
```

```python
b = a
b[0] = 333
b
```

```python
a
```

Matrix Operations

```python
np.dot(A,B)
```

Advance Array Manipulation 
 stacking array

```python
a1 = np.array([1,2,3,4,5,6,7,8])
a2 = np.array([9,10,11,12,13,14,15,16])
#row wise stack
np.vstack((a1 , a2))
```

```python
#column wise 
np.hstack((a1,a2))
```

```python
np.column_stack((a1, a2))
```
