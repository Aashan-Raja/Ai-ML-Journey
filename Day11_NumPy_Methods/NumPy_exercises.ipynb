```python
import numpy as np
#Columns: [Age, Math Marks, Science Marks]
data = np.array([
                 [18,85,78],
                 [19,92,88],
                 [17,76,95],
                 [18,65,70],
                 [20,90,85]
                 ])
```

Get the shape of matrix.

```python
data.shape
```

Find the average age of students.

```python
avg = np.mean(data, axis = 0)
print("Average age : " , avg[0])
```
