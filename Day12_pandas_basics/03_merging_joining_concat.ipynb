```python
import numpy as np 
import pandas as pd
```

**Merging two Data Frames** 

```python
#dataframe 1:Employees information
employes = pd.DataFrame({
    'employee_id' : [1,2,3,4,5],
    'name' :['John','Anna','Peter','Linda','Bob'],
    'department' :['HR','IT','Finance', 'IT','HR']
})
#dataframe 2: Salary information
salaries = pd.DataFrame({
    'employee_id' :[1,2,3,6,7],
    'salary' :[1000,1222,2345,5435,453],
    'bonus' :[233,232,54,66,64]
}
)
```

```python
employes
```

```python
salaries
```

```python
pd.merge(employes,salaries ,on='employee_id',how="inner")
```

```python
pd.merge(employes,salaries ,on='employee_id',how="outer")
```

```python
pd.merge(employes,salaries ,on='employee_id',how="left")
```

```python
pd.merge(employes,salaries ,on='employee_id',how="right")
```

**Concatenation**

```python
employes
```

```python
salaries
```

```python
pd.concat([employes,salaries], axis=0, ignore_index= True)
```

**Joining**

```python
df1  = pd.DataFrame({
    'name' : ['Ali' , 'Hamza' , 'Zain']
}
)
df2 = pd.DataFrame({
    'marks' :[10,20,30,40]
})
```

```python
df1.join(df2 , how='outer')
```
