```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A' :[1,2,np.nan,4,5],
    'B' :[np.nan,2,3,4,5],
    'C' :[1,2,3,np.nan,np.nan],
    'D' :[1,np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df
```

```python
df.dropna(axis=0)
```

```python
df.dropna(thresh=3)
```

Filling  the missing Data

```python
df
```

```python
#filling missing values with 0
df.fillna(0)
```

```python
#fill missing values with different values for each column
values = {'A':100,'B':1200,'C':200 ,'D':21}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
